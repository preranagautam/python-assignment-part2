```python
#---------------------------------------- Part 2: Data Structures -----------------------------------------

import copy
import datetime
import warnings
warnings.filterwarnings("ignore")

#Given Data Structures 
menu = {
    "Paneer Tikka":   {"category": "Starters",  "price": 180.0, "available": True},
    "Chicken Wings":  {"category": "Starters",  "price": 220.0, "available": False},
    "Veg Soup":       {"category": "Starters",  "price": 120.0, "available": True},
    "Butter Chicken": {"category": "Mains",     "price": 320.0, "available": True},
    "Dal Tadka":      {"category": "Mains",     "price": 180.0, "available": True},
    "Veg Biryani":    {"category": "Mains",     "price": 250.0, "available": True},
    "Garlic Naan":    {"category": "Mains",     "price":  40.0, "available": True},
    "Gulab Jamun":    {"category": "Desserts",  "price":  90.0, "available": True},
    "Rasgulla":       {"category": "Desserts",  "price":  80.0, "available": True},
    "Ice Cream":      {"category": "Desserts",  "price": 110.0, "available": False},
}

inventory = {
    "Paneer Tikka":   {"stock": 10, "reorder_level": 3},
    "Chicken Wings":  {"stock":  8, "reorder_level": 2},
    "Veg Soup":       {"stock": 15, "reorder_level": 5},
    "Butter Chicken": {"stock": 12, "reorder_level": 4},
    "Dal Tadka":      {"stock": 20, "reorder_level": 5},
    "Veg Biryani":    {"stock":  6, "reorder_level": 3},
    "Garlic Naan":    {"stock": 30, "reorder_level": 10},
    "Gulab Jamun":    {"stock":  5, "reorder_level": 2},
    "Rasgulla":       {"stock":  4, "reorder_level": 3},
    "Ice Cream":      {"stock":  7, "reorder_level": 4},
}

sales_log = {
    "2025-01-01": [
        {"order_id": 1,  "items": ["Paneer Tikka", "Garlic Naan"],          "total": 220.0},
        {"order_id": 2,  "items": ["Gulab Jamun", "Veg Soup"],              "total": 210.0},
        {"order_id": 3,  "items": ["Butter Chicken", "Garlic Naan"],        "total": 360.0},
    ],
    "2025-01-02": [
        {"order_id": 4,  "items": ["Dal Tadka", "Garlic Naan"],             "total": 220.0},
        {"order_id": 5,  "items": ["Veg Biryani", "Gulab Jamun"],           "total": 340.0},
    ],
    "2025-01-03": [
        {"order_id": 6,  "items": ["Paneer Tikka", "Rasgulla"],             "total": 260.0},
        {"order_id": 7,  "items": ["Butter Chicken", "Veg Biryani"],        "total": 570.0},
        {"order_id": 8,  "items": ["Garlic Naan", "Gulab Jamun"],           "total": 130.0},
    ],
    "2025-01-04": [
        {"order_id": 9,  "items": ["Dal Tadka", "Garlic Naan", "Rasgulla"], "total": 300.0},
        {"order_id": 10, "items": ["Paneer Tikka", "Gulab Jamun"],          "total": 270.0},
    ],
}

#----------------------------------------- Task 1 — Explore the Menu ------------------------------------------
categories = list({uc["category"] for uc in menu.values() if "category" in uc})     #extracting unique categories from menu

for category in categories:                                                         #printing menu category-wise
  print(f"\n{'========'} {category} {'========'}") 
  for item_name, details in menu.items():
    if details["category"]==category:
      status="[Available]" if details["available"] else "[Unavailable]"
      print(f"{item_name:<16}  ₹{details['price']:<8}   {status}")
#printing menu statistics
total_items_count = len(menu)
avlbl_items_count = len({item_name for item_name, details in menu.items() if details["available"]})
most_expensive_item = max(menu, key=lambda item_name: menu[item_name]["price"])
most_expensive_price=menu[most_expensive_item]["price"]
print(f"\n{'Total number of Items:':<28} {total_items_count}")
print(f"{'Number of Available Items:':<28} {avlbl_items_count}")
print(f"{'Most Expensive Item:':<28} {most_expensive_item} ₹{most_expensive_price}\n")
print(f"{'Items under ₹150.00:':^30} \n-------------------------------")
for item_name, details in menu.items():                                                 #menu items under ₹150
  if details["price"]<150:
    print(f"{item_name:<20} \t ₹{details['price']:<8}")

#--------------------------------------------- Task 2: Cart Operations -----------------------------------------------
cart = []                                                                               #initialising empty cart

#method to find an item in the cart and return its entry (dictionary) if found, else None
def find_in_cart(cart,item_name):
  for entry in cart:
    if entry["item"]==item_name:
      return entry
  return None

#method to add an item to the cart, or update quantity if it already exists. Also checks menu availability and existence.
def add_to_cart(cart,item_name,quantity):
  if item_name not in menu:
    print(f" ✗ {item_name} is not on the menu.")
    return
  if not menu[item_name]["available"]:
    print(f" ✗ {item_name} is currently unavailable.")
    return
  #logic to update quantity if item already in cart, else add new entry
  it_exists = find_in_cart(cart,item_name)
  if it_exists:
    it_exists["quantity"]+=quantity
    print(f"✓ Updated '{item_name}' x new quantity: {it_exists['quantity']}")
  else:
    cart.append({"item": item_name, "quantity": quantity, "price": menu[item_name]["price"]})
    print(f"✓ Item Added '{item_name}' x quantity: {quantity}")

#method to remove an item from the cart if it exists, else print not found message
def remove_from_cart(cart,item_name):
  it_exists=find_in_cart(cart, item_name)
  if it_exists:
    cart.remove(it_exists)
    print(f"✓ Removed '{item_name}' from cart.")
  else:
    print(f"✗ '{item_name}' not found in cart.")

"""def update_quantity(cart,item_name,new_quantity):
  it_exists=find_in_cart(cart, item_name)
  if it_exists:
    it_exists["quantity"]=new_quantity
    print(f"✓ Updated '{item_name}' x new quantity: {new_quantity}.")
  else:
    print(f"✗ '{item_name}' not found in cart.")"""

#method to print the current contents of the cart in a formatted manner, showing item name, quantity and line total
def print_cart(cart):
  print(f"\n{'='*8} Shopping Cart {'='*8}")
  if not cart:
    print("Cart is empty.")
    return
  for entry in cart:
    print(f"{entry['item']:<18} x{entry['quantity']}"
          f"\t₹{entry['quantity'] * entry['price']:.2f}")

#method to print the order summary with line totals, subtotal, GST and total payable amount
def print_order_summary(cart):
  print(f"\n\n{'='*10} Order Summary {'='*10}") 
  subtotal = 0 
  for entry in cart:
    line_total = entry["quantity"] * entry["price"]
    subtotal += line_total
    print(f"{entry['item']:<18} x {entry['quantity']} \t ₹{line_total:.2f}")
  gst = round(subtotal * 0.05, 2)
  total = round(subtotal + gst, 2)
  print(f"{'-' * 35}")
  print(f"{'Subtotal:':<18} \t ₹{subtotal:.2f}") 
  print(f"{'GST (5%):':<18} \t ₹{gst}")
  print(f"{'Total Payable:':<18} \t ₹{total}")
  print(f"{'='*35}")

# Demonstrating cart operations
print(f"\n\n")
print(f"\n---------- Add Paneer Tikka x 2 ------------")
add_to_cart(cart, "Paneer Tikka", 2)
print_cart(cart)

print(f"\n----------- Add Gulab Jamun x 1 ------------")
add_to_cart(cart, "Gulab Jamun", 1)
print_cart(cart)

print(f"\n---------- Add Paneer Tikka x 1 ------------")
add_to_cart(cart, "Paneer Tikka", 1)
print_cart(cart)

print(f"\n------ Try to add Mystery Burger x 1 -------")
add_to_cart(cart, "Mystery Burger", 1)
print_cart(cart)

print(f"\n------ Try to add Chicken Wings x 1 --------")
add_to_cart(cart, "Chicken Wings", 1)
print_cart(cart)

print(f"\n----------- Remove Gulab Jamun -------------")
remove_from_cart(cart, "Gulab Jamun")
print_cart(cart)

print_order_summary(cart)

#---------------------------------- Task 3: Inventory Tracker with Deep Copy ---------------------------------
inventory_backup = copy.deepcopy(inventory)                                     #creating a deep copy of inventory to preserve original data

#manually change one stock value in inventory 
print(f"\n\n--------- Demonstrating Deep Copy works ------------")
inventory["Garlic Naan"]["stock"] = 100
print(f"\nInventory: {inventory}")                                                #print both inventory and inventory_backup to show that the backup is unaffected
print(f"\nInventory Backup: {inventory_backup}")
inventory["Garlic Naan"]["stock"] = inventory_backup["Garlic Naan"]["stock"]    #restore inventory
print(f"\nInventory restored.\n")

print(f"\n--------- Deducting Stock for fulfilled orders ------------")
for entry in cart:
  item_name = entry["item"]
  qty_needed = entry["quantity"]
  current = inventory[item_name]["stock"]
#logic to check if current stock is sufficient to fulfill the needed quantity, if yes deduct needed quantity, else set stock to 0 and print warning
  if current >= qty_needed:
    inventory[item_name]["stock"] -= qty_needed
    print(f"✓ {qty_needed} {item_name}(s) fulfilled.")
    print(f"\n {item_name}: {current} --> {inventory[item_name]["stock"]}")
  else:
    warnings.warn(f"✗ Insufficient stock for {item_name}.\nCurrent quantity = {current} \t Needed quantity = {qty_needed} \nDeducting all available.")
    inventory[item_name]["stock"] = 0

print(f"\n--------- Reorder Alerts ------------")
alerted = False
#logic to check if any item's stock has fallen to or below its reorder level, if yes print a reorder alert message
for item_name, details in inventory.items():
  if details["stock"] <= details["reorder_level"]:
    print(f"\n⚠ Reorder Alert: {item_name} - Only {details["stock"]} unit(s) left. (reorder level: {details["reorder_level"]})")
    alerted = True
if not alerted:
  print("\n✓ No reorder alerts.")
#print inventory and backup again to show that inventory has been altered while backup is unchanged
print(f"\nCurrent (altered) Inventory: {inventory}")#['Paneer Tikka']['stock']}")
print(f"\nInventory Backup (unaltered): {inventory_backup}")

#----------------------------------------------- TASK 4 - Daily Sales Log Analysis----------------------------------------------------
revenue_per_day={}                                                              #initialising empty dictionary to store revenue per day
print(f"\n\n--------- Revenue Per Day ------------")
for date, orders in sales_log.items():
    revenue_per_day[date] = sum(order['total'] for order in orders)
    print(f"  {date} : ₹{revenue_per_day[date]:.2f}")

bestselling_day = max(revenue_per_day, key=lambda d: revenue_per_day[d])        #finding the day with maximum revenue
print(f"\n  Best-selling day: {bestselling_day} - ₹{revenue_per_day[bestselling_day]:.2f}")

#------------------------- Most ordered item ---------------------------------------------------
item_count={}
for date, orders in sales_log.items():
    for order in orders:
        for item in order['items']:
          item_count[item] = item_count.get(item, 0) + 1
most_ordered_item = max(item_count, key=lambda item: item_count[item])
print(f"\n Most ordered item: {most_ordered_item}, appeared in {item_count[most_ordered_item]} orders.")

#------------------------- Updating sales log with new orders and printing updated stats -------------------------
sales_log["2025-01-05"] = [
    {"order_id": 11, "items": ["Butter Chicken", "Gulab Jamun", "Garlic Naan"], "total": 490.0},
    {"order_id": 12, "items": ["Paneer Tikka", "Rasgulla"],                     "total": 260.0},
]
print("\n--- Updated Revenue Per Day ---")
new_rev_per_day = {}
for date, orders in sales_log.items():
    new_rev_per_day[date] = sum(order["total"] for order in orders)
    print(f"  {date} : ₹{new_rev_per_day[date]:.2f}")

new_best_day = max(new_rev_per_day, key=lambda d: new_rev_per_day[d])
print(f"\n  Updated best-selling day: {new_best_day} — ₹{new_rev_per_day[new_best_day]:.2f}")

#--------------------------- Numbered list of all orders across all dates ---------------------------
print("\n--- All Orders (Numbered) ---")
counter = 1
for date, orders in sales_log.items():
    for order in orders:
        items_str = ", ".join(order["items"])
        print(f"  {counter:<3} [{date}] Order #{order['order_id']:<3}"
              f" — ₹{order['total']:.2f} — Items: {items_str}")
        counter += 1

#--------------------------------------- End of Part 2: Data Structures -----------------------------------------
```
